```python
#creating random tensors
#in notes give example of nn first initialises with random weights then sees examples and updates representation outputs and tweak its random value acc to examples and then repeat with more examples
#creating 2random (but the same) tensors

random_1= tf.random.Generator.from_seed(42) #set seed for reproducibility
random_1=random_1.normal(shape=(3,2))
random_2= tf.random.Generator.from_seed(42)
random_2=random_2.normal(shape=(3,2))

#Are they equal?
random_1,random_2 , random_1==random_2
#not exactly random but pseudorandom cos they have seeds
```

```python
#shuffle a tensor (valuable when you want to shuffle your data so the inherent learning is not affected)
#in notes give example like if there was food image classification neural networkand we had 15k imgs of sphaghetti , 10k of ramen , then nn will see sphagetti first and then its pattern will be too alligned to spahaghetti img than to a ramen img
not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])
#shuffle our non-shuffled tensor
tf.random.shuffle(not_shuffled)
#shuffles it along its 1st dimension which means 3 , not 2
```

```python
#it looks like if we want out shuffled tensors to be in the same order , we' ve got to use the global lvl random seed and operational level random seed
tf.random.set_seed(42) #global lvl random seed
tf.random.shuffle(not_shuffled, seed=42)#oprtnl lvl random seed , if used alone , will get different results each time
```

```python
#getting information from tensors
#shape , rank , axis or dimn , size

#creating a rank 4 tensor (4 dimensions)
rank4_tensor= tf.zeros(shape=[2,3,4,5])
rank4_tensor
```

```python
rank4_tensor[0]
```

```python
rank4_tensor.shape , rank4_tensor.ndim , tf.size(rank4_tensor)
```

```python
#get various attributes of our tensor
print("Datatype of every element:", rank4_tensor.dtype)
print("Number of dimensions (rank):", rank4_tensor.ndim)
print("Shape of tensor:", rank4_tensor.shape)
print("Elements along the 0 axis:", rank4_tensor.shape[0])
print("Elements along the last axis:", rank4_tensor.shape[-1])
print("Total number of elements in our tensor:", tf.size(rank4_tensor))
print("Total number of elements in our tensor:", tf.size(rank4_tensor).numpy())
```

```python
# indexing tensors
#tensors can be indexed just like python lists
some_list= [1,2,3,4]
some_list[:2]

#get the first 2 elements of each dimension
rank4_tensor[:2,:2,:2,:2]
```

```python
#get 1st element from each dimension from each index except for the final one
rank4_tensor[:1,:1,:1,:]
```

```python
#create a rank 2 tensor
rank2_tensor=tf.constant([[10,7],
                          [3,4]])
rank2_tensor
```

```python
#get last item of each row
rank2_tensor[:,-1]  #get every row , last column
```

```python
#add in extra dimension to our rank 2 tensor
rank3_tensor = rank2_tensor[..., tf.newaxis] #... means :,: (every access beforr that)
rank3_tensor
```

```python
#alternate to tf.newaxis
tf.expand_dims(rank2_tensor, axis=-1) #'-1 means expand the final axis
```

```python
tf.expand_dims(rank2_tensor , axis=0) #expand the 0 axis
```

```python
tf.expand_dims(rank2_tensor, axis=1) #expand the 1 axis
```
